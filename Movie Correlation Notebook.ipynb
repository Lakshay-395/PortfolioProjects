```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df=pd.read_csv(r'C:\Users\Lakshay\Downloads\movies.csv') 
```

```python
df.head()
```

```python
#Check missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(pct_missing)
```

```python
#Data types for columns
df.dtypes
```

```python
df['budget']= df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df = df.fillna(0)
```

```python
df.dtypes
```

```python
df
```

```python
df['Year_released'] = df['released'].astype('string')
```

```python
df
```

```python
df.drop('Year_released',axis=1)
```

```python
df = df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
df.head()
```

```python
#finding correlations with revenue
```

```python
#Scatter plot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings ')
plt.xlabel('Gross earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
#Budget vs Gross plot using Seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'green'})
```

```python
#Correlation
df.corr(method='pearson')
```

```python
#High correlation b/w budget and gross
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Features ')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
